```python
from transformers import AutoTokenizer, AutoModel, pipeline
from transformers import BertTokenizer, BertModel
from transformers import pipeline
```

```python
generator = pipeline(task="text-generation", model="distilgpt2")
text="this is very amazing"
output=generator(text)
output
```

```python
text="Messi is the best player in the football history"
output=generator(text,num_return_sequences=2, return_full_text=True)
output
```

```python
classifier=pipeline("sentiment-analysis")
text="this is very amazing"
output=classifier(text)
output
```

```python
classifier=pipeline("zero-shot-classification")
text="I love messi goals"
output=classifier(text,candidate_labels=['education','football','culture'])
output
```

```python
text="I love messi goals and matches"
output=classifier(text,candidate_labels=['education','football','agriculture'])
output
```

```python
ner = pipeline("ner", grouped_entities=True)
text = "Barack Obama was born in Hawaii and served as President of the United States."
output = ner(text)
output
```
